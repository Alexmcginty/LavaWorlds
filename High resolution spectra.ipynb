```
#Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```
#Import the Mercury data for fo2=0 high res
path = 'Basalt_1995_fo2_0_high_res_flux_ratio.dat'
if 'flux_ratio' in path:
        data = np.loadtxt(path)
        lamda_basa, spec_basa = data[:,0], data[:,1]*1e6 #Specfiy the wavelength and spectral data
#Import the Komatiite data for fo2=0 high res
path1 = 'Komatiite_1995_fo2_0_high_res_flux_ratio.dat'
if 'flux_ratio' in path:
        data1 = np.loadtxt(path1)
        lamda_koma, spec_koma = data1[:,0], data1[:,1]*1e6 #Specfiy the wavelength and spectral data
```

```
fig, axes = plt.subplots(figsize=(10,6),nrows=1,ncols=2) #Create the figure
ax = axes[0] #Set axis
ax.plot(lamda_basa,spec_basa) #Plot the Basalt spectrum
ax.set_xlim(0.7,0.8) #Set xlim
ax.set_title('Basalt') #Set title
ax1 = axes[1] #Set axis
ax1.plot(lamda_koma,spec_koma) #Plot the Komatiite spectrum
ax1.set_xlim(0.7,0.8) #Set xlim
ax1.set_title('Komatiite') #Set title
fig.text(0.5,0.05,'Wavelength ($\mu$m)', va='center', ha='center',fontsize='x-large')
fig.text(0.05,0.5,'F$_p$/F$_*$ ppm', rotation='vertical',va='center', ha='center',fontsize='x-large')
plt.savefig('high res spectrum fo2=0.png')
```
