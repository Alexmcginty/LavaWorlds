```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Import the chemistry data as pandas data frame
df = pd.read_csv('chemistry_Basalt.dat',delimiter ='\t')
df1 = pd.read_csv('chemistry_Komatiite.dat',delimiter ='\t')
df2 = pd.read_csv('chemistry_BSE_-2.dat',delimiter ='\t')
df3 = pd.read_csv('chemistry_BSE_2.dat',delimiter ='\t')
```

```python
#Stpre the Atomic and Molecular mixing ratios from columns in the dataframe
Oxygen_mixing = df.iloc[:,6]
O2_mixing= df.iloc[:,29]
Mg_mixing= df.iloc[:,7]
MgO_mixing= df.iloc[:,24]
SiO_mixing= df.iloc[:,28]
SiO2_mixing= df.iloc[:,30]
Na_mixing= df.iloc[:,11]
NaO_mixing= df.iloc[:,26]
K_mixing= df.iloc[:,12]
Fe_mixing= df.iloc[:,13]
FeO_mixing= df.iloc[:,21]

Oxygen_mixing_koma = df1.iloc[:,6]
O2_mixing_koma= df1.iloc[:,27]
Mg_mixing_koma= df1.iloc[:,7]
MgO_mixing_koma= df1.iloc[:,22]
SiO_mixing_koma= df1.iloc[:,26]
SiO2_mixing_koma= df1.iloc[:,28]
Na_mixing_koma= df1.iloc[:,11]
NaO_mixing_koma= df1.iloc[:,24]
K_mixing_koma= df1.iloc[:,12]
```

```python
fig, ax = plt.subplots(figsize=(8,6)) #Create figure
#Plot each abudance profile for Basalt
ax.plot(Oxygen_mixing,df.iloc[:,0]*1e-6,label='O')
ax.plot(O2_mixing,df.iloc[:,0]*1e-6,label='O2')
ax.plot(Mg_mixing,df.iloc[:,0]*1e-6,label='Mg')
ax.plot(MgO_mixing,df.iloc[:,0]*1e-6,label='MgO')
ax.plot(SiO_mixing,df.iloc[:,0]*1e-6,label='SiO')
ax.plot(SiO2_mixing,df.iloc[:,0]*1e-6,label='SiO2')
ax.plot(Na_mixing,df.iloc[:,0]*1e-6,label='Na')
ax.plot(NaO_mixing,df.iloc[:,0]*1e-6,label='NaO')
ax.plot(K_mixing,df.iloc[:,0]*1e-6,label='K')
ax.plot(Fe_mixing,df.iloc[:,0]*1e-6,label='Fe')
ax.plot(FeO_mixing,df.iloc[:,0]*1e-6,label='FeO',color='black')
#Customise the axis
ax.set_yscale('log')
ax.set_xscale('log')
ax.invert_yaxis()
ax.set_xlabel('Mixing Ratios')
ax.set_ylabel('Pressure [bars]')
ax.legend() #Add a legend
```

```python
fig, ax = plt.subplots(figsize=(8,6)) #Create figure
#Plot the abudence profiles for Komatiite
ax.plot(Oxygen_mixing_koma,df1.iloc[:,0]*1e-6,label='O')
ax.plot(O2_mixing_koma,df1.iloc[:,0]*1e-6,label='O2')
ax.plot(Mg_mixing_koma,df1.iloc[:,0]*1e-6,label='Mg')
ax.plot(MgO_mixing_koma,df1.iloc[:,0]*1e-6,label='MgO')
ax.plot(SiO_mixing_koma,df1.iloc[:,0]*1e-6,label='SiO')
ax.plot(SiO2_mixing_koma,df1.iloc[:,0]*1e-6,label='SiO2')
ax.plot(Na_mixing_koma,df1.iloc[:,0]*1e-6,label='Na')
ax.plot(NaO_mixing_koma,df1.iloc[:,0]*1e-6,label='NaO')
ax.plot(K_mixing_koma,df1.iloc[:,0]*1e-6,label='K')
#Customise the axis
ax.set_yscale('log')
ax.set_xscale('log')
ax.invert_yaxis()
ax.set_xlabel('Mixing Ratios')
ax.set_ylabel('Pressure [bars]')
ax.legend() #Add a legend
```
