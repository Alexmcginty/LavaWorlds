```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
import matplotlib as mpl
```

I downloaded the data in from glamdring and read the data in below to the correct arrays. There is definitly a more elgenat way of doing this but i just did a rough method which worked.
For the height coordinates I plotted the height for the highest fo2 for each compostion. This was to give non exoplanet people a more intuitive way of understanding the vertical space.

```python
#Load the Basalt TP data in
Basalt_fo2_m2 = 'Basalt_1995_fo2_-2_tp.dat'
data1 = np.loadtxt(Basalt_fo2_m2,skiprows=2,usecols=(1,2,3))
BASA_fo2_m2_temp, BASA_fo2_m2_press , BASA_fo2_m2_h = data1[:,0], data1[:,1] , data1[:,2]
Basalt_fo2_m1 = 'Basalt_1995_fo2_-1_tp.dat'
data2 = np.loadtxt(Basalt_fo2_m1,skiprows=2,usecols=(1,2))
BASA_fo2_m1_temp, BASA_fo2_m1_press = data2[:,0], data2[:,1]
Basalt_fo2_0 = 'Basalt_1995_fo2_0_tp.dat'
data3 = np.loadtxt(Basalt_fo2_0,skiprows=2,usecols=(1,2))
BASA_fo2_0_temp, BASA_fo2_0_press = data3[:,0], data3[:,1]
Basalt_fo2_1 = 'Basalt_1995_fo2_1_tp.dat'
data4 = np.loadtxt(Basalt_fo2_1,skiprows=2,usecols=(1,2))
BASA_fo2_1_temp, BASA_fo2_1_press = data4[:,0], data4[:,1]
Basalt_fo2_2 = 'Basalt_1995_fo2_2_tp.dat'
data5 = np.loadtxt(Basalt_fo2_2,skiprows=2,usecols=(1,2))
BASA_fo2_2_temp, BASA_fo2_2_press = data5[:,0], data5[:,1]
#Load the BSE TP data in
BSE_fo2_m2 = 'BSE_1995_fo2_-2_tp.dat'
data6 = np.loadtxt(BSE_fo2_m2,skiprows=2,usecols=(1,2,3))
BSE_fo2_m2_temp, BSE_fo2_m2_press, BSE_fo2_m2_h = data6[:,0], data6[:,1], data6[:,2]
BSE_fo2_m1 = 'BSE_1995_fo2_-1_tp.dat'
data7 = np.loadtxt(BSE_fo2_m1,skiprows=2,usecols=(1,2))
BSE_fo2_m1_temp, BSE_fo2_m1_press = data7[:,0], data7[:,1]
BSE_fo2_0 = 'BSE_1995_fo2_0_tp.dat'
data8 = np.loadtxt(BSE_fo2_0,skiprows=2,usecols=(1,2))
BSE_fo2_0_temp, BSE_fo2_0_press = data8[:,0], data8[:,1]
BSE_fo2_1 = 'BSE_1995_fo2_1_tp.dat'
data9 = np.loadtxt(Basalt_fo2_1,skiprows=2,usecols=(1,2))
BSE_fo2_1_temp, BSE_fo2_1_press = data9[:,0], data9[:,1]
BSE_fo2_2 = 'BSE_1995_fo2_2_tp.dat'
data10 = np.loadtxt(BSE_fo2_2,skiprows=2,usecols=(1,2))
BSE_fo2_2_temp, BSE_fo2_2_press = data10[:,0], data10[:,1]
#Load the Komatiite data in
Komatiite_fo2_m2 = 'Komatiite_1995_fo2_-2_tp.dat'
data11 = np.loadtxt(Komatiite_fo2_m2,skiprows=2,usecols=(1,2,3))
Komatiite_fo2_m2_temp, Komatiite_fo2_m2_press, Komatiite_fo2_m2_h = data11[:,0], data11[:,1] , data11[:,2]
Komatiite_fo2_m1 = 'Komatiite_1995_fo2_-1_tp.dat'
data12 = np.loadtxt(Komatiite_fo2_m1,skiprows=2,usecols=(1,2))
Komatiite_fo2_m1_temp, Komatiite_fo2_m1_press = data12[:,0], data12[:,1]
Komatiite_fo2_0 = 'Komatiite_1995_fo2_0_tp.dat'
data13 = np.loadtxt(Komatiite_fo2_0,skiprows=2,usecols=(1,2))
Komatiite_fo2_0_temp, Komatiite_fo2_0_press = data13[:,0], data13[:,1]
Komatiite_fo2_1 = 'Komatiite_1995_fo2_1_tp.dat'
data14 = np.loadtxt(Komatiite_fo2_1,skiprows=2,usecols=(1,2))
Komatiite_fo2_1_temp, Komatiite_fo2_1_press = data14[:,0], data14[:,1]
Komatiite_fo2_2 = 'Komatiite_1995_fo2_2_tp.dat'
data15 = np.loadtxt(Komatiite_fo2_2,skiprows=2,usecols=(1,2))
Komatiite_fo2_2_temp, Komatiite_fo2_2_press = data15[:,0], data15[:,1]
#Load the mercury data in
Mercury_fo2_m2 = 'Mercury_1995_fo2_-2_tp.dat'
data16 = np.loadtxt(Mercury_fo2_m2,skiprows=2,usecols=(1,2,3))
Mercury_fo2_m2_temp, Mercury_fo2_m2_press, Mercury_fo2_m2_h = data16[:,0], data16[:,1], data16[:,2]
Mercury_fo2_m1 = 'Mercury_1995_fo2_-1_tp.dat'
data17 = np.loadtxt(Mercury_fo2_m1,skiprows=2,usecols=(1,2))
Mercury_fo2_m1_temp, Mercury_fo2_m1_press = data17[:,0], data17[:,1]
Mercury_fo2_0 = 'Mercury_1995_fo2_0_tp.dat'
data18 = np.loadtxt(Mercury_fo2_0,skiprows=2,usecols=(1,2))
Mercury_fo2_0_temp, Mercury_fo2_0_press = data18[:,0], data18[:,1]
Mercury_fo2_1 = 'Mercury_1995_fo2_1_tp.dat'
data19 = np.loadtxt(Mercury_fo2_1,skiprows=2,usecols=(1,2))
Mercury_fo2_1_temp, Mercury_fo2_1_press = data19[:,0], data19[:,1]
Mercury_fo2_2 = 'Mercury_1995_fo2_2_tp.dat'
data20 = np.loadtxt(Mercury_fo2_2,skiprows=2,usecols=(1,2))
Mercury_fo2_2_temp, Mercury_fo2_2_press = data20[:,0], data20[:,1]
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(14,10),sharey=True,sharex=True) #Create figure
h = np.linspace(0,1189.59,49)
#Create a color map to use when plotting various fo2s
cmap = plt.get_cmap('magma',5)
cm=plt.cm.magma(np.linspace(0,1,5))
ax=axes[0,0] #Set axis
#Plot the Basalt TP profiles
ax.plot(BASA_fo2_m2_temp, BASA_fo2_m2_press*1e-6,label='fo2=-2 TP profile',color=cm[0])
ax.plot(BASA_fo2_m1_temp, BASA_fo2_m1_press*1e-6,label='fo2=-1 TP profile',color=cm[1])
ax.plot(BASA_fo2_0_temp, BASA_fo2_0_press*1e-6,label='fo2=0 TP profile',color=cm[2])
ax.plot(BASA_fo2_1_temp, BASA_fo2_1_press*1e-6,label='fo2=1 TP profile',color=cm[3])
ax.plot(BASA_fo2_2_temp, BASA_fo2_2_press*1e-6,label='fo2=2 TP profile',color=cm[4])
ax.set(xlim=[2750,3050],ylim=[1e-2,1e-8],yscale='log') #Customise axis
ax.set_title('Basalt TP plots') #Title the plot
Ax =ax.twinx() #Create a secondary axis
#Here a T-height plot was created which wasn't shown in the figure to have a second y axis
Ax.plot(BASA_fo2_m1_temp,BASA_fo2_m2_h[0:49]/(1000*100),label='',color='white',linestyle='') 
ax.legend(loc='best', frameon=True) #Add legend

ax1=axes[1,0] #Set axis
#Plot the BSE TP profiles
ax1.plot(BSE_fo2_m2_temp, BSE_fo2_m2_press*1e-6,label='BSE 1995K fo2=-2 TP profile',color=cm[0])
ax1.plot(BSE_fo2_m1_temp, BSE_fo2_m1_press*1e-6,label='BSE 1995K fo2=-1 TP profile',color=cm[1])
ax1.plot(BSE_fo2_0_temp, BSE_fo2_0_press*1e-6,label='BSE 1995K fo2=0 TP profile',color=cm[2])
ax1.plot(BSE_fo2_1_temp, BSE_fo2_1_press*1e-6,label='BSE 1995K fo2=1 TP profile',color=cm[3])
ax1.plot(BSE_fo2_2_temp, BSE_fo2_2_press*1e-6,label='BSE 1995K fo2=2 TP profile',color=cm[4])
ax1.set(xlim=[2750,3050],ylim=[1e-2,1e-8], yscale='log') #Customise axis
ax1.set_title('BSE TP plots') #Title the plot
Ax1 =ax1.twinx() #Create a secondary axis
#Here a T-height plot was created which wasn't shown in the figure to have a second y axis
Ax1.plot(BSE_fo2_m1_temp,BSE_fo2_m2_h[0:44]/(1000*100),label='',color='white',linestyle='')

ax2=axes[0,1] #Set axis
#Plot the Mercury TP profiles
ax2.plot(Mercury_fo2_m2_temp, Mercury_fo2_m2_press*1e-6,label='Mercury 1995K fo2=-2 TP profile',color=cm[0])
ax2.plot(Mercury_fo2_m1_temp, Mercury_fo2_m1_press*1e-6,label='Mercury 1995K fo2=-1 TP profile',color=cm[1])
ax2.plot(Mercury_fo2_0_temp, Mercury_fo2_0_press*1e-6,label='Mercury 1995K fo2=0 TP profile',color=cm[2])
ax2.plot(Mercury_fo2_1_temp, Mercury_fo2_1_press*1e-6,label='Mercury 1995K fo2=1 TP profile',color=cm[3])
ax2.plot(Mercury_fo2_2_temp, Mercury_fo2_2_press*1e-6,label='Mercury 1995K fo2=2 TP profile',color=cm[4])
ax2.set(xlim=[2750,3050],ylim=[1e-2,1e-8], yscale='log')
ax2.set_title('Mercury TP plots') #Title the plot
Ax2 =ax2.twinx() #Create a secondary axis
#Here a T-height plot was created which wasn't shown in the figure to have a second y axis
Ax2.plot(Mercury_fo2_m1_temp,Mercury_fo2_m2_h[0:40]/(1000*100),label='',color='white',linestyle='')

ax3=axes[1,1] #Set axis
#Plot the Komatiite TP profiles
ax3.plot(Komatiite_fo2_m2_temp, Komatiite_fo2_m2_press*1e-6,label='Komatiite 1995K fo2=-2 TP profile',color=cm[0])
ax3.plot(Komatiite_fo2_m1_temp, Komatiite_fo2_m1_press*1e-6,label='Komatiite 1995K fo2=-1 TP profile',color=cm[1])
ax3.plot(Komatiite_fo2_0_temp, Komatiite_fo2_0_press*1e-6,label='Komatiite 1995K fo2=0 TP profile',color=cm[2])
ax3.plot(Komatiite_fo2_1_temp, Komatiite_fo2_1_press*1e-6,label='Komatiite 1995K fo2=1 TP profile',color=cm[3])
ax3.plot(Komatiite_fo2_2_temp, Komatiite_fo2_2_press*1e-6,label='Komatiite 1995K fo2=2 TP profile',color=cm[4])
ax3.set(xlim=[2750,3050],ylim=[1e-2,1e-8], yscale='log')
Ax3 =ax3.twinx()
Ax3.plot(Komatiite_fo2_m1_temp,Komatiite_fo2_m2_h[0:43]/(1000*100),label='',color='white',linestyle='')
#ax3.legend(loc='best', frameon=True)
ax3.set_title('Komatiite TP plots')
fig.text(0.5,0.05,'Temperature [K]', va='center', ha='center',fontsize='x-large')
fig.text(0.08,0.5,'Pressure [bar]', rotation='vertical',va='center', ha='center',fontsize='x-large')
fig.text(0.95,0.5,'Height [km]', rotation='vertical',va='center', ha='center',fontsize='x-large')
plt.savefig('TP summary plots.png')
```
